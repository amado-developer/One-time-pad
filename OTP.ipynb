```python
#OTP with just text
#Amado Garcia 181469
import random

charset = "ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"


def main():
    vector = "Me Muero de sueenio"
    encrypted = encrypt(vector)
    decrypted = decrypt(encrypted[0], encrypted[1])

    print("Test Vector: " + vector)
    print("OTP: " + encrypted[0])
    print("Encrypted: " + encrypted[1])
    print("Decrypted: " + decrypted)


def encrypt(plaintext):
    """Encrypt plaintext value.
    Keyword arguments:
    plaintext -- the plaintext value to encrypt.
    """
    otp = "".join(random.sample(charset, len(charset)))
    result = ""

    for c in plaintext.upper():
        if c not in otp:
            continue
        else:
            result += otp[charset.find(c)]

    return otp, result


def decrypt(otp, secret):
    """Decrypt secret value.
    Keyword arguments:
    otp -- the one-time pad used when the secret value was encrypted.
    secret -- the value to be decrypted.
    """
    result = ""

    for c in secret.upper():
        if c not in otp:
            continue
        else:
            result += charset[otp.find(c)]

    return result


if __name__ == "__main__":
    main()
```

```python
#OTP asking for files
#Amado Garcia 181469
#Code is based in https://www.tutorialspoint.com/cryptography_with_python/cryptography_with_python_implementation_of_one_time_pad_cipher.htm#:~:text=Python%20includes%20a%20hacky%20implementation,one%2Dtime%20pad%20cipher%20algorithm.
#https://www.learnpythonwithrune.org/understand-the-security-of-one-time-pad-and-how-to-implement-it-in-python/

from random import randint
charset = 'aAbBcCdDeEfFgGhHiIjJkKlLmMnNoOpPqQrRsStTuUvVwWxXyYzZ,./\;:"[]{}()_-!@#$%^&*|?<>`~'

def encrypt(plaintext, filen):
    txt = ''
    for p,c in enumerate(plaintext):
        if c not in charset:
            txt += c
        else:
            encrypted = (charset.index(c) + int(filen[p])) % 81
            txt += charset[encrypted]
    return txt

def decrypt(txt, filen):
    plaintext = ''
    for p,c in enumerate (txt):
        if c not in charset:
            plaintext += c
        else:
            decrypted = (charset.index(c) - int(filen[p])) % 81
            plaintext += charset[decrypted]

    return plaintext

def first(filename):
    with open(filename, "r") as f:
        contents = f.read().splitlines()
    return contents

def create_files_otp(s_file, length):
    for x in range(s_file):
        with open("otp" + str(x) + ".txt", "w") as f:
            for i in range (length):
                f.write(str(randint(0, 81)) + "\n")

def save_file(filename, data):
    with open(filename, 'w') as f:
        f.write(data)

def get_plaintext():
    plain_text = input('Enter your message: ')
    return plain_text

def load_file(filename):
    with open(filename, 'r') as f:
        contents = f.read()
    return contents

def menu():
    flag = "1"

    while flag == "1":
        print('Option 1. Create OTP')
        print('Option 2. Encrypt message')
        print('Option 3. Decrypt message')
        print('Option 4. Exit program')
        choice = input('Choose your number option: ')

        if choice == '1':
            otp_files = int(input('How many OTP do you want '))
            length = int(input('Maximum message length '))
            create_files_otp(otp_files, length)

        elif choice == '2':
            filename = input('Filename of the OTP you want: ')
            otp_files = first(filename)
            plaintext = get_plaintext()
            txt = encrypt(plaintext, otp_files)
            filename = input('Name of output encrypted file: ')
            save_file(filename, txt)

        elif choice == '3':
            filename = input('Filename of the OTP you: ')
            otp_files = first(filename)
            filename = input('Output file name (decrypted): ')
            txt = load_file(filename)
            plaintext = decrypt(txt, otp_files)
            print('The decrypted message is:'+ plaintext)

        elif choice == '4':
            print("Bye")
            flag = "2"
        
        else:
            print("Invalid option \n")

menu()
```
